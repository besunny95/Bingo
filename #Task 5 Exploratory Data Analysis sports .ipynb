### maximum wickets taken by bowlers

```python
p['dismissal_kind'].unique()
```

```python
dismissd= p[p.dismissal_kind != 'run out']                 ## these two are not considers as the bowlers wickets.
dis=dismissd[dismissd.dismissal_kind != 'retired hurt']
```

```python
dis.info()
```

```python
t_dis = dis.groupby('bowler').apply(lambda x:x ['dismissal_kind'].dropna()).reset_index(name='wickets')
c_wkts = t_dis.groupby('bowler').count().reset_index()
a= c_wkts.sort_values(by='wickets',ascending=False)
topbow = a.loc[:,['bowler','wickets']][0:10]
topbow
```
